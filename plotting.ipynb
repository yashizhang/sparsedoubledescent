```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Using matplotlib, draw a plot depicting two graphs: 
# (1) the bias variance tradeoff U curve, and
# (2) the double descent curve. 
# The x-axis should be the number of parameters in the model, and the y-axis should be the expected risk.

# Bias-Variance Tradeoff
x = np.linspace(0, 1, 100)
y = (x-0.5)**2 
# change color of line
plt.figure(figsize=(15, 5))
plt.plot(x, y, color='orange')
x = np.linspace(1, 2.5, 100)
y = np.exp(-x) - (1/np.e) + 1/4
plt.plot(x, y, color='orange')
# increase label font size
plt.xlabel('Model Complexity', fontsize=25) 
plt.ylabel('Risk', fontsize=25)
plt.yticks([])
plt.xticks([1])
# increase tick font size
plt.tick_params(axis='both', which='major', labelsize=25)
# remove top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# change figure size to be horizontal


plt.show()
```
